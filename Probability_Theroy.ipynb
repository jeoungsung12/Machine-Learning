```python
import numpy as np
rng = np.random.default_rng(seed=2022)
rng.integers(low=0, high=3, size=10)

rng1 = np.random.default_rng(seed=2022)
rng1.integers(low=0, high=3, size=10)
rng2 = np.random.default_rng(seed=2022)
rng2.integers(low=0, high=3, size=10)
```

```python
rng.random(3) # Generate random numbers in [0, 1) 닫힌구간, 열린구간
rng1.random(3) # Generate random numbers in [0, 1) 닫힌구간, 열린구간
rng2.random(3) # Generate random numbers in [0, 1) 닫힌구간, 열린구간
```

```python
3*rng.random((2,3)) + 2 # Generate random numbers in [2, 5)
3*rng1.random((2,3)) + 255 #Gaussian noise1(mu = 0, sigma^2 = 0.1)
3*rng2.random((2,3)) + 255 #Gaussian noise1(mu = 0, sigma^2 = 0.5)
```

```python
mu, sigma = 0, 0.1 # Mean and standard deviation 
s = rng.normal(mu, sigma, 10000)
print(np.mean(s))
print(np.std(s, ddof=1))
```

```python
import matplotlib.pyplot as plt
count, x, ignored = plt.hist(s, 50, density=True)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (x - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
#Gaussian noise1(mu = 0, sigma^2 = 0.1)
mu, sigma = 0, np.sqrt(0.1) # Mean and standard deviation
img = plt.imread("IMG_1476.jpg") # Change to your own picture
noise = np.random.normal(mu, sigma, (710, 710, 3)) 
img = np.uint8(((img/255)+noise)*255)
s = rng.normal(mu, sigma, 10000)
print(np.mean(s))
print(np.std(s, ddof=1))
import matplotlib.pyplot as plt
count, x, ignored = plt.hist(s, 50, density=True)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (x - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
import matplotlib.pyplot as plt
print(img.shape) # This will show the size of your image (height, width, channel)
plt.imshow(img) #Gaussian noise1(mu = 0, sigma^2 = 0.1)
plt.show()
```
